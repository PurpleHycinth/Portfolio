```python
import pandas as pd

df1 = pd.read_excel('KIDB-export-29457ae0-8a89-4b0a-a8d7-f0a940ea3ffb.xlsx')
df2 = pd.read_excel('KIDB-export-d7648f85-207a-4459-b61e-6740ffc9bba2.xlsx')
df1.head()
```

```python
df2= df2.iloc[:,:23]

# Melt the DataFrame
melted_df = pd.melt(df2, id_vars=['Economy', 'Indicator', 'Unit of Measure'], var_name='Year', value_name='Value')

# Optionally, you can convert the 'Year' column to numeric if needed
melted_df['Year'] = pd.to_numeric(melted_df['Year'], errors='coerce')

# Save the melted DataFrame to a new file
melted_df.to_csv('melted_dataset2.csv', index=False)
```

```python
import dash
from dash import dcc, html
import plotly.express as px
import pandas as pd
from dash import Output, Input

# Assuming 'melted_dataset.csv' is the file generated in the previous step
df = melted_df.copy()

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    html.H1("Domestic Material Consumption by Year and Country"),
    
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': country, 'value': country} for country in df['Economy'].unique()],
        value=df['Economy'].iloc[0],
        multi=False,
        style={'width': '50%'}
    ),
    
    dcc.Graph(id='line-chart'),
])

# Callback to update the line chart based on the selected country
@app.callback(
    Output('line-chart', 'figure'),
    [Input('country-dropdown', 'value')]
)
def update_chart(selected_country):
    filtered_df = df[df['Economy'] == selected_country]
    fig = px.line(filtered_df, x='Year', y='Value', title=f'Domestic Material Consumption for {selected_country}')
    return fig

# Run the app in the notebook
app.run_server(debug=False)
```
